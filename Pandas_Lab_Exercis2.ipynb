```python
import pandas as pd
```

**Load the Automobile dataset into variable "auto"**

```python
df = pd.read_csv('Automobile.csv')
df
```

**Check the head of the DataFrame.**

```python
df.head()
```

**How many rows and columns are there?**

```python
print(df.shape)
print(df.info())
```

**What is the average Price of all cars in the dataset?**

```python
ap = df['price'].mean()
print(f"The average price of all cars is: {ap}")
```

**Which is the cheapest make and costliest make of car in the lot?**

```python
cm = df['price'].min()
print(f"The cheapest make of car in the lot is: {cm}")
```

**How many cars have horsepower greater than 100?**

```python
hp = df[df['horsepower']>100]
print(f"Cars with Horsepower greater than 100:{len(hp)}")
print(hp)
```

**How many hatchback cars are in the dataset ?**

```python
hb = df[df['body_style']=='hatchback']
print(f"The number of hatchback cars in the dataset is: {len(hb)}")
print(hb)
```

**What are the 3 most commonly found cars in the dataset?**

```python
cc = df['make'].value_counts().nlargest(3)
print(f"The 3 most commonly found cars are:\n{cc}")
```

**Someone purchased a car for 7099, what is the make of the car?**

```python
mc = df[df['price']==7099]
print(f"The make of the car is: {mc['make']}")
```

**Which cars are priced greater than 40000?**

```python
gt = df[df['price']>40000]
print(f"The cars priced greater than 40000 are:\n{gt}")
```

**Which are the cars that are both a sedan and priced less than 7000?**

```python
sp = df[(df['body_style']=='sedan') & (df['price']<7000)]
print(f"The cars that are both a sedan and priced less than 7000 are:\n{sp}")
```

 **Count the number of unique values in the `fuel_type` column.**

```python
u = df['fuel_type'].unique()
k = df['fuel_type'].nunique()
print(f"The number of unique values in the fuel_type column is: {k}")
print(f"The two unique elements in the fuel_type column are: {u}")
```

**List all the cars that have a horsepower between 100 and 200, and display their `make`, `horsepower`, and `price`.**

```python
ca = df[(df['horsepower']>=100) & (df['horsepower']<=200)]
print(ca[['make','horsepower','price']])
```

**Find the average `city_mpg` and `highway_mpg` for each `body_style`.**

```python
cav = df.groupby('body_style')['city_mpg'].mean()
print(cav)
```

```python
hav = df.groupby('body_style')['highway_mpg'].mean()
print(hav)
```

**What is the median `price` for each `make`?**

```python
m = df.groupby('make')['price'].median()
print(m)
```

**List all cars that have a `wheel_base` greater than 100 and a `curb_weight` less than 2500.**

```python
lc = df[(df['wheel_base']>100) & (df['curb_weight']<2500)]
print(lc)
```

**Create a new column `price_per_hp` that calculates the price of the car per horsepower.**

```python
df['price_per_hp'] = df['price']/df['horsepower']
df
```

**Count how many cars have a `number_of_doors` as `four`.**

```python
f = df[df['number_of_doors']=='four'].value_counts()
print(len(f))
```

**Find the top 5 cars based on their `highway_mpg` and `price`.**

```python
tp = df.sort_values(by=['highway_mpg','price'],ascending=False).head(5)
print(tp)
```

**How many cars have missing values in the `normalized_losses` column?**

```python
df.isnull().sum()
```

**Create a new column `car_age` that calculates the age of the car based on the `year_of_manufacture` (assume the current year is 2025).**

```python
import numpy as np
current_year = 2025
base_year = 2000
max_price = df['price'].max()
df['year_of_manufacture'] = np.round(base_year + (df['price'] / max_price) * (current_year - base_year))
df['car_age'] = current_year - df['year_of_manufacture']
print(df[['make', 'price', 'year_of_manufacture', 'car_age']].head())
```
